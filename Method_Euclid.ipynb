```python
def EuclidSub(a, b):
    while (a != b):
        if a > b:
            k  = a - b
            a = k
        else:
            k = b - a
            b = k

    return a
```

```python
EuclidSub(24, 18)
```

```python
def SieveOfEratosthenes(N):
    arr = [i for i in range(1, N + 1)]
    arr_del = [0 for i in range(1, N + 1)]
    
    for i in range(1, N):
        for j in range(i + 1, N):
            if arr_del[j] == 0 and arr[j] % arr[i] == 0:
                arr_del[j] = 1

    res = []
    for i in range(0, N):
        if arr_del[i] == 0:
            res.append(arr[i])
            
    return res
```

```python
prime_arr = SieveOfEratosthenes(1000)

print(prime_arr)
print(len(prime_arr))
```
