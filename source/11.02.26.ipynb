Решение квадратного уравнения
$$ ax^2 + bx + c = 0 $$

```python
import math
```

```python
def SolvingQuadEq(a: float, b: float, c: float):
    D = b**2 - 4 * a * c
    #print(D)
    if D == 0:
        x = -b / (2 * a)
        return ((x, 0), (x, 0))
    if D > 0:
        x_1 = (-b + math.sqrt(D)) / (2 * a)
        x_2 = (-b - math.sqrt(D)) / (2 * a)
        return ((x_1, 0), (x_2, 0))
    else:
        re = -b / (2 * a)
        im_1 = math.sqrt(abs(D)) / (2 * a)
        im_2 = -math.sqrt(abs(D)) / (2 * a)
        return ((re, im_1), (re, im_2))
```

```python
a = 1
b = -1
c = 1

r = SolvingQuadEq(a, b, c)
for i in (0, 1):
    print(f"{r[i][0]} + {r[i][1]}i")
```

```python
def EqualZero(number) -> bool:
    if abs(number) < 0.0001:
        return True
    else:
        return False
```

```python
def DeterminantDuo(a_1, b_1, a_2, b_2):
    return a_1 * b_2 - b_1 *  a_2
```

```python
def MethodKramer(a_1, b_1, c_1, a_2, b_2, c_2):
    Delta = DeterminantDuo(a_1, b_1, a_2, b_2)
    
    delta_x = DeterminantDuo(c_1, b_1, c_2, b_2)
    delta_y = DeterminantDuo(a_1, c_1, a_2, c_2)
    
    return (delta_x / Delta, delta_y / Delta)
```

```python
def MethodGauss(a_1, b_1, c_1, a_2, b_2, c_2):
    g_1 = 1
    d_1 = b_1 / a_1
    e_1 = c_1 / a_1
    g_2 = 0
    d_2 = b_2 - a_2 * d_1
    e_2 = c_2 - a_2 * e_1
    
    if EqualZero(d_2) and EqualZero(e_2):
        return (f'{e_1 - d_1} * y', 'any')
    elif EqualZero(d_2) and not EqualZero(e_2):
        return ('no solution', 'no solution')
```

```python
def SolvingLinearSystem(a_1, b_1, c_1, a_2, b_2, c_2):
    det = DeterminantDuo(a_1, b_1, a_2, b_2)
    if not EqualZero(det):
        r = MethodKramer(a_1, b_1, c_1, a_2, b_2, c_2)
        print(f"x = {r[0]}\ny = {r[1]}")
    
    else:
        r = MethodGauss(a_1, b_1, c_1, a_2, b_2, c_2)
        print(f"x = {r[0]}\ny = {r[1]}")
```

```python
a_1 = 1
b_1 = -2
c_1 = 8
a_2 = 2
b_2 = 1
c_2 = 1

SolvingLinearSystem(a_1, b_1, c_1, a_2, b_2, c_2)
```
